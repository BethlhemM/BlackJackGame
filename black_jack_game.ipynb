```python
import random
import card_display as cd # use cd.card_display(list_cards) for image display

"""
   A class used to generate a deck of 52 poker card names in string type
   
"""  
class Card:
    def __init__ (self, suit, val):
        self.suit = suit
        self.value = val

    def show(self):
        print ("{} _of_ {}".format(self.value,self.suit))

class Deck:
    deck = []
    def __init__ (self):
        #self.cards = []
        self.build()
        
    def build(self):
        for i in ["spades", "clubs", "diamonds", "hearts"]:
            for j in range (1,14):
                if j == 1:
                    self.deck.append("{}_of_{}".format("ace",i))
                elif j==11:
                    self.deck.append("{}_of_{}".format("jack",i))
                elif j==12:
                    self.deck.append("{}_of_{}".format("queen",i))
                elif j==13:
                    self.deck.append("{}_of_{}".format("king",i))
                else:
                    self.deck.append("{}_of_{}".format(j,i))
    def show(self):
        for c in self.deck:
            print(c)
            
    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self):
        return self.deck.pop()

    def size():
        return len(self.deck)







```

```python
# class Player
import card_display as cd 

# Player class
class Player:
    def __init__(self):
        self.cards = []
        self.points = 0
        
    def add_card(self,card):
        self.cards.append(card)
        card = card.split("_")[0]
        
        if card == "ace":
            self.points+=11
        elif card in ["jack","queen","king"]:
            self.points += 10
        else:
            self.points +=int(card)
        
    def display(self):     
        cd.card_display(self.cards)




```

```python
# Play Game for two cards only: 

if __name__=="__main__":
    deck = Deck()
    deck.shuffle()

    card1 = deck.deal()
    card2 = deck.deal()
    card3 = deck.deal()
    card4 = deck.deal()

    player = Player()
    dealer = Player()

    player.add_card(card1)
    player.add_card(card2)

    dealer.add_card(card3)
    dealer.add_card(card4)

    
        
    print("For player:",end=" ")
    player.display()

    print("For dealer:",end=" ")
    dealer.display()

    if player.points > dealer.points:
        print("Player won!")
    elif player.points < dealer.points:
        print("Dealer won!")
    else:
        print("They tie the game!")
    
    print("Player has {} points, Dealer has {} points".format(player.points,dealer.points))






```
